#### Plot both $y=x^2$ and $y=e^x$ in the same plot using normal and logarithmic scale.
**Hint**: Use `set_xscale` and `set_yscale`.

```python

# Assuming x, y1, and y2 are defined
x = np.linspace(0, 5, 100)
y1 = x**2
y2 = np.exp(x)

# Create a figure and axis
fig, ax = plt.subplots()

# Plot y = x^2 in normal scale
ax.plot(x, y1, label='y = x^2', color='blue', linestyle='-', linewidth=2)

# Plot y = e^x in normal scale
ax.plot(x, y2, label='y = e^x', color='red', linestyle='--', linewidth=2)

# Adding labels and title
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_title('Plot of y = x^2 and y = e^x')

# Adding legend
ax.legend()

# Display the plot with normal scale
plt.show()

# Create a new figure and axis for logarithmic scale
fig, ax_log = plt.subplots()

# Plot y = x^2 in logarithmic scale
ax_log.plot(x, y1, label='y = x^2', color='blue', linestyle='-', linewidth=2)

# Plot y = e^x in logarithmic scale
ax_log.plot(x, y2, label='y = e^x', color='red', linestyle='--', linewidth=2)

# Adding labels and title
ax_log.set_xlabel('x-axis (log scale)')
ax_log.set_ylabel('y-axis (log scale)')
ax_log.set_title('Plot of y = x^2 and y = e^x (logarithmic scale)')

# Adding legend
ax_log.legend()

# Set logarithmic scale on both axes
ax_log.set_xscale('log')
ax_log.set_yscale('log')

# Display the plot with logarithmic scale
plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
# Create a figure and axis
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
                         
# Plot 1: Median steps by month
plt.sca(axes[0])
fitbit.groupby('Months_encoded')['Steps'].median().plot(kind='bar', color='skyblue')
plt.title('Median Steps by Month')
plt.xlabel('Month (Encoded)')
plt.ylabel('Median Steps')

# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
# Plot 2: Median steps by Work or Weekend
plt.sca(axes[1])
fitbit.groupby('Work_or_Weekend')['Steps'].median().plot(kind='bar', color='lightgreen')
plt.title('Median Steps by Weekday or Weekend')
plt.xlabel('Weekday or Weekend')
plt.ylabel('Median Steps')
plt.xticks([1, 0], ['Work', 'Weekend'])


# Save figures to the 'figures' folder
plt.savefig('figures/fitbit_visualizations.png')

# Show the plots
plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here

# Features to plot
features = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']

# Create a figure with subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Loop through features and create scatter plots
for i, feature in enumerate(features):
    plt.sca(axes[i])
    plt.scatter(fitbit[feature], fitbit['Steps'], alpha=0.5)
    plt.title(f'{feature} vs Steps')
    plt.xlabel(feature)
    plt.ylabel('Steps')

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show the plots
plt.show()
```

#### Use the right visulalization to show the distribution of column `Age`.

```python
# your code here
# Set the plot style to classic and the figure size
plt.style.use('classic')
plt.figure(figsize=(12, 6))

# Plot the distribution of 'Age' using a histogram
sns.histplot(data=titanic, x='Age', bins=30, kde=True, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

#### Use subplots and plot the distribution of the `Age`  with bins equal to 10, 20 and 50.

```python
# your code here

# Set the plot style to classic and the figure size
plt.style.use('classic')
plt.figure(figsize=(18, 6))

# Define the number of bins
bins_values = [10, 20, 50]

# Create subplots
for i, bins in enumerate(bins_values, 1):
    plt.subplot(1, 3, i)
    sns.histplot(data=titanic, x='Age', bins=bins, kde=True, color='skyblue')
    plt.title(f'Distribution of Age (Bins={bins})')
    plt.xlabel('Age')
    plt.ylabel('Frequency')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

#### How does the bin size affect your plot?

```python
"""
The bin size in a histogram affects the granularity with which the data is presented. Smaller bin sizes provide a more detailed view of the distribution, allowing you to see individual data points more clearly. On the other hand, larger bin sizes result in a smoother and more generalized representation of the distribution.
"""
```

#### We would like to have in one single plot the summary statistics of the feature `Age`. What kind of plot would you use? Plot it. 

```python
# your code here
# Set the style and figure size
sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))

# Create a box plot for the Age feature
sns.boxplot(x=titanic['Age'])

# Set plot labels and title
plt.title('Summary Statistics of Age')
plt.xlabel('Age')

# Show the plot
plt.show()
```

```python
"""
This box plot will display the median (line inside the box), quartiles (box), and potential outliers (points outside the whiskers) for the Age feature.
"""
```

#### What does the last plot tell you about the feature `Age`?

```python
q1 = titanic['Age'].quantile(0.25)
q2 = titanic['Age'].median()
q3 = titanic['Age'].quantile(0.75)

print("Q1:", q1)
print("Q2 (Median):", q2)
print("Q3:", q3)
```

#### We suspect that there is a linear relationship between `Fare` and `Age`. Use the right plot to show the relationship between these 2 features. There are 2 ways, please do it both ways.

```python
# Method 1 - matplotlib
# your code here

# Create a scatter plot using Matplotlib
plt.figure(figsize=(10, 6))
plt.scatter(titanic['Age'], titanic['Fare'], alpha=0.5, color='blue')
plt.title('Relationship between Age and Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```

```python
# Method 2 - seaborn
# your code here

# Create a scatter plot using Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Fare', data=titanic, alpha=0.5, color='green')
plt.title('Relationship between Age and Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```

#### Plot the correlation matrix using seaborn.

```python
# your code here

numeric_columns = titanic.select_dtypes(include='number')

# Handle missing values if needed
numeric_columns = numeric_columns.dropna()

# Create correlation matrix
correlation_matrix = numeric_columns.corr()

# Create a heatmap using seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix')
plt.show()

```

#### What are the most correlated features?

```python
"""
The "pclass" is represented by an ascending numerical order and a descending  order of quality; that is, the lower the "pclass" number, the better it is in terms of quality. Therefore, there is a strong negative correlation between "pclass" and "fare," as the lower the class number, the better the quality and the more expensive it will be. Additionally, there is a negative correlation between age and "pclass," indicating that older individuals chose higher-quality classes. Finally, there is a weak relationship between survival and "fare." While we know there is a strong relationship between the better classes and the fare paid, we cannot extrapolate that the survival rate is directly linked to the chosen class.

"""
```

#### Use seaborn to plot the distribution of `Age` based on the `Gender`.
**Hint**: Use Facetgrid.

```python
# your code here

sns.set(style="whitegrid")

# Create a FacetGrid with Seaborn
g = sns.FacetGrid(titanic, col="Gender", height=6, aspect=1.2)

# Map the distribution plot onto the grid
g.map(sns.histplot, 'Age', bins=20, kde=True, color='skyblue')

# Set plot labels and title
g.set_axis_labels("Age", "Count")
g.set_titles(col_template="{col_name} Gender")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```
